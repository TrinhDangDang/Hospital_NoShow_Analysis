```python
import pandas as pd

df = pd.read_csv("KaggleV2-May-2016.csv")
df.head()
df.info()
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```

```python
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days
df = df[df['waiting_days'] >= 0]  # remove any invalid negative wait times
```

```python
df['no_show'] = df['no_show'].map({'No': 0, 'Yes': 1})
```

```python
no_show_rate = df['no_show'].mean()
print(f"No-show rate: {no_show_rate:.2%}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='gender', hue='no_show')
plt.title('No-show Rate by Gender')
plt.show()
```

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 60, 100], labels=['0-18', '19-35', '36-60', '60+'])

sns.countplot(data=df, x='age_group', hue='no_show')
plt.title('No-show Rate by Age Group')
plt.show()
```

```python
df.to_csv("cleaned_noshow_data.csv", index=False)
```

```python
sms_group = df.groupby('sms_received')['no_show'].mean()
print("No-show rate with/without SMS:\n", sms_group)
```

```python
no_show_by_neighborhood = df.groupby('neighbourhood')['no_show'].mean().sort_values(ascending=False).head(10)
print("Top 10 neighborhoods by no-show rate:\n", no_show_by_neighborhood)
```

```python
df.to_csv("final_cleaned_noshow_data.csv", index=False)
```
